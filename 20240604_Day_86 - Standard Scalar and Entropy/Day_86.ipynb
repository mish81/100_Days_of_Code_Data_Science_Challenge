```python
import numpy as np
import pandas as pd

def entropy(x):
    return np.round(-np.sum(x * np.log2(x)), 4)

df = pd.DataFrame(
    {
        'val_1': np.arange(0.01, 1.0, 0.1),
        'val_2': 1 - np.arange(0.01, 1.0, 0.1),
    }
)

df['entropy'] = df.apply(lambda row: entropy([row['val_1'], row['val_2']]), axis=1)

print(df)
```
