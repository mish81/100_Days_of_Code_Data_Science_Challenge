```python
import pandas as pd

google = pd.read_csv('google.csv', index_col=0)
google = google.reset_index()
google['Date'] = pd.to_datetime(google['Date'])
google['Year'] = pd.DatetimeIndex(google['Date']).year
google['Month'] = pd.DatetimeIndex(google['Date']).month 

print(google)
```
