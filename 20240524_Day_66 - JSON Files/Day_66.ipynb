```python
import pandas as pd
```

```python
df = pd.read_json('market.json')
```

```python
df.head()
```

```python
df['pct_change'] = df['pct_change'].str.replace('%)','')
df['pct_change'] = df['pct_change'].str.replace('(','')
df['pct_change'] = df['pct_change'].astype(float)
print(df)
```

```python
pd.set_option('display.max_columns', 15)
pd.set_option('display.width', 1000)
df = pd.read_json('market.json')
df['pct_change'] = df['pct_change'].apply(lambda item: float(item[1:-2]))
```

```python
print(df.loc[df['pct_change'].sort_values(ascending=False).head(1).index])
print()
print(df.loc[df['pct_change'].sort_values(ascending=True).head(1).index])
```

```python
print(df.iloc[df['pct_change'].argmax()])
print()
print(df.iloc[df['pct_change'].argmin()])
```

```python
pd.set_option('display.max_columns', 15)
pd.set_option('display.width', 1000)
df = pd.read_json('market.json')
df['pct_change'] = df['pct_change'].apply(lambda item: float(item[1:-2]))
```

```python
df[df['pct_change']>=0].to_json('positive.json')
df[df['pct_change']<0].to_json('negative.json')
```
