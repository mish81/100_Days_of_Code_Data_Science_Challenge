```python
import numpy as np
import pandas as pd

np.random.seed(42)
s1 = pd.Series(np.random.rand(20))
s2 = pd.Series(np.random.randn(20))

df = pd.concat([s1, s2], axis=1)
df.columns = ['col1', 'col2']

df['col3'] = df['col2'].apply(lambda x : 1 if x >= 0 else -1)

print(df)
```

```python
df['col4'] = df['col2'].apply(lambda x : -1 if x <= -1 else (1 if x >= 1 else x))
print(df)
```

```python
df2 = df.sort_values(by='col2',ascending=False)
df3 = df.sort_values(by='col2',ascending=True)
print(df2['col2'].head())
print()
print(df3['col2'].head())
```
