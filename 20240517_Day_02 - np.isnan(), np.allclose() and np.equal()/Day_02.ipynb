Basic Usage

np.isnan(a) checks each element of the array a and returns a boolean array indicating whether each element is NaN.

Parameters

a: Input array or object that can be converted to an array.

```python
import numpy as np
```

```python
A = np.array([[3, 2, 1, np.nan],
              [5, np.nan, 1, 6]])
```

```python
print(np.isnan(A))
```

Basic Usage

np.allclose(a, b) checks if all elements of array a are approximately equal to the corresponding elements of array b within a specified tolerance.

Parameters

a, b: Input arrays to be compared.

rtol (optional): The relative tolerance parameter (default is 1e-05). It represents the tolerance relative to the values in b.

atol (optional): The absolute tolerance parameter (default is 1e-08). It represents the minimum absolute tolerance level.

equal_nan (optional): If True, NaNs in a will be considered equal to NaNs in b.

```python
A = np.array([0.4, 0.5, 0.3])
B = np.array([0.39999999, 0.5000001, 0.3])
```

```python
print(np.allclose(A,B))
```

```python
A = np.array([0.4, 0.5, 0.3])
B = np.array([3.9999999, 5.000001, 0.3])
```

```python
print(np.allclose(A,B))
```

Basic Usage

np.equal(a, b) performs element-wise comparison of the arrays a and b.

Parameters

a, b: Input arrays to be compared. They should have the same shape or be broadcastable to a common shape.

```python
A = np.array([0.4, 0.5, 0.3])
B = np.array([0.3999999999, 0.5000000001, 0.3])
```

```python
print(np.equal(A,B))
```

```python
A = np.array([0.4, 0.5, 0.3, 4, np.nan])
B = np.array([0.3999999999, 0.5000000001, 0.3])
```

```python
# Both arrays need to be of the same size
print(np.equal(A,B))
```

```python
# Comparison with scalar
a = np.array([1, 2, 3])
result = np.equal(a, 2)
print(result)
```
