```python
import numpy as np
```

```python
A = np.array([[5, 8, 16],
              [4, 1, 8],
              [-4, 4, -11]])
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
print(eigenvalues)
print(eigenvectors)
```
