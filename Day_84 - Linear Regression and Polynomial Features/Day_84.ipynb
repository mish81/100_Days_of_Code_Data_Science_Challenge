```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.DataFrame(
    {
        'years': [1, 2, 3, 4, 5, 6],
        'salary': [4000, 4250, 4500, 4750, 5000, 5250],
    }
)
```

```python
# Feature Matrix
X = np.array(df['years']).reshape(-1, 1)

# Target vector
y = np.array(df['salary']).reshape(-1, 1)

# Fitting Linear Regression Model from sklearn
reg = LinearRegression().fit(X,y)

print(f"Linear regression: {reg.intercept_[0]:.2f} + {reg.coef_[0][0]:.2f}x")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_csv("data.csv")
```

```python
model = LinearRegression()
reg = model.fit(df[['variable']],df[['target']])
```

```python
# Calculate the model score
score = model.score(df[['variable']], df[['target']])
print(round(score,4))
```
