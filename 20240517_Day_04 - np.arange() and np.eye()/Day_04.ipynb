```python
import numpy as np
```

Basic Usage

np.arange([start,] stop[, step,], dtype=None)

Parameters
- start (optional): Start of the interval. The default is 0.
- stop: End of the interval. The array will not include this value.
- step (optional): Step size between values. The default is 1.
- dtype (optional): The data type of the output array. If not specified, the data type is inferred from other input arguments.

```python
print(np.arange(10,100))
```

Basic Usage

ndarray.reshape(shape, order='C')

Parameters
- shape: The new shape of the array. This can be specified as a tuple of integers or as separate integer arguments (e.g., reshape(rows, columns) for a 2D array).
- order (optional): The order in which to read/write elements in the reshaped array. 'C' means to read/write the elements using C-style indexing (row-major), and 'F' means to use Fortran-style indexing (column-major).

```python
data = np.arange(10,100)
print(data.reshape([9,10]))
```

Basic Usage

np.eye(N, M=None, k=0, dtype=float)

Parameters
- N: Number of rows in the output matrix.
- M (optional): Number of columns in the output matrix. If not specified, defaults to N.
- k (optional): Index of the diagonal. The default is 0 (main diagonal). Positive values shift the diagonal up, and negative values shift it down.
- dtype (optional): The data type of the output array. Default is float.

```python
print(np.eye(6,6,k=0,dtype=int))
```
