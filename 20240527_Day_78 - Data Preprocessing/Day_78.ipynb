```python
import pandas as pd

df = pd.DataFrame(
    data={
        'hashtags': [
            '#good#vibes',
            '#hot#summer#holiday',
            '#street#food',
            '#workout',
        ]
    }
)
```

```python
help(pd.Series.str.split)
```

```python
df = df['hashtags'].str.split('#',expand=True)
df.columns = ['del','hashtag1','hashtag2','hashtag3']
df.drop(['del'],axis=1,inplace=True)
print(df)
```

```python
df['missing'] = df.isna().sum(axis=1)
print(df)
```

```python
df = pd.DataFrame(
    data={
        'investments': [
            '100_000_000',
            '100_000',
            '30_000_000',
            '100_500_000',
        ]
    }
)
```

```python
df['investments'] = df['investments'].str.replace('_','').astype(int)
print(df)
```
