```python
import numpy as np
```

```python
A = np.linspace(0, np.pi/2, 20)
B = np.full(shape=(20,), fill_value=1, dtype=float)
```

```python
print(A)
print(B)
```

```python
# Calculating sin and cos of A
sin_A = np.sin(A)
cos_A = np.cos(A)

# Calculating sin and cos squared of A
sin_2A = np.square(sin_A)
cos_2A = np.square(cos_A)

# Adding squared value of sin and cos to get Pythagorean Identity
A_sin_cos = sin_2A + cos_2A

# Comparison against B
print(np.allclose(A_sin_cos,B))
```
