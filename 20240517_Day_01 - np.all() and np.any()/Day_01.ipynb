What np.all(B) Does:

np.all(B) checks whether all elements in the entire array B are True (non-zero).

```python
import numpy as np
```

```python
A = np.array([[3, 2, 1, 4],
              [5, 2, 1, 6]])
 
B = np.array([[3, 2, 1, 4],
              [5, 2, 0, 6]])
 
C = np.array([[True, False, False],
              [True, True, True]])
 
D = np.array([0.1, 0.3])
```

```python
print(np.all(A))
print(np.all(B))
print(np.all(C))
print(np.all(D))
```

To check for a particular axis, variable "axis = " can be added to the code.

```python
import numpy as np
```

```python
A = np.array([[3, 2, 1, 4],
              [5, 2, 1, 6]])
 
B = np.array([[3, 2, 1, 4],
              [5, 2, 0, 6]])
 
C = np.array([[True, False, False],
              [True, True, True]])
 
D = np.array([0.1, 0.3])
```

```python
print(np.all(A,axis = 1))
print(np.all(B,axis = 1))
print(np.all(C,axis = 1))
```

Basic Usage

np.any(a) returns True if at least one element of the array a is True. Otherwise, it returns False.

Parameters

a: Input array or object that can be converted to an array.

axis (optional): The axis or axes along which to perform the operation. By default, it checks the entire array.

np.any(a, axis=0) checks each column.

np.any(a, axis=1) checks each row.

```python
import numpy as np
```

```python
A = np.array([[0, 0, 0],
              [0, 0, 0]])

B = np.array([[0, 0, 0],
              [0, 1, 0]])

C = np.array([[False, False, False],
              [True, False, False]])

D = np.array([[0.1, 0.0]])
```

```python
print(np.any(A))
print(np.any(B))
print(np.any(C))
print(np.any(D))
```
