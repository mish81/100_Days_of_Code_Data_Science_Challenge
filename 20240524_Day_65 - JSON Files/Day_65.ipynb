```python
import pandas as pd
```

```python
df = pd.read_json('market.json')
print(df)
```

```python
import re

pattern = r'\((.*?)\)'

df['country'] = df['profile'].apply(lambda x : re.search(pattern,x).group(1))
print(df)
```

```python
country_count = df['country'].value_counts()
sorted_data = sorted(country_count[country_count > 1].index)
print(sorted_data)
```
