```python
import numpy as np
```

```python
coeff_1 = np.array([2,4,-5,1])
coeff_2 = np.array([2,0,-5,1])
```

```python
print(np.roots(coeff_1))
print(np.roots(coeff_2))
```

```python
coeff_1 = np.array([4,5,1])
coeff_2 = np.array([2,4,-5,1])
```

```python
help(np.polyadd)
```

```python
print(np.polyadd(coeff_1,coeff_2))
print(np.polysub(coeff_1,coeff_2))
print(np.polymul(coeff_1,coeff_2))
print(np.polyadd(coeff_1,2*coeff_2))
```
