```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split


pd.set_option('display.max_columns', 10)
pd.set_option('display.width', 200)
np.set_printoptions(precision=2, suppress=True, linewidth=100)
raw_data = load_breast_cancer()

data = raw_data['data']
target = raw_data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data, target, random_state=40, test_size=0.25, stratify=target
)

for name, array in zip(['target', 'y_train', 'y_test'], [target, y_train, y_test]):
    print(
        f'{name.ljust(7)}:'
        f'{np.unique(array, return_counts=True)[1] / len(array)}'
    )
```

```python
import numpy as np
import pandas as pd


df = pd.DataFrame(
    {
        'years': [1, 2, 3, 4, 5, 6],
        'salary': [4000, 4250, 4500, 4750, 5000, 5250],
    }
)
```

```python
# Feature matrix (add a column of ones for the intercept)
X = np.vstack([np.ones(len(df['years'])), df['years']]).T

# Target vector
y = df['salary'].values

# Normal equation: theta = (X.T * X)^(-1) * X.T * y
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

# Extract coefficients
intercept = theta[0]
slope = theta[1]

print(f"Linear regression: {intercept:.2f} + {slope:.2f}x")
```
