```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

data = {
    'size': ['XL', 'L', 'M', np.nan, 'M', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red', 'green'],
    'gender': ['female', 'male', np.nan, 'female', 'female', 'male'],
    'price': [199.0, 89.0, np.nan, 129.0, 79.0, 89.0],
    'weight': [500, 450, 300, np.nan, 410, np.nan],
    'bought': ['yes', 'no', 'yes', 'no', 'yes', 'no']
}

df = pd.DataFrame(data=data)
```

```python
df_object = df.select_dtypes('object')
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value='empty')
df_object[list(df_object.columns)] = imputer.fit_transform(df_object[list(df_object.columns)])
print(df_object)
```

```python
import pandas as pd

df = pd.DataFrame(data={'weight': [75., 78.5, 85., 91., 84.5, 83., 68.]})
```

```python
help(pd.cut)
```

```python
df['weight_cut'] = pd.cut(df['weight'],bins=3)
print(df)
```

```python
df = pd.DataFrame(data={'weight': [75., 78.5, 85., 91., 84.5, 83., 68.]})
```

```python
bins = pd.IntervalIndex.from_tuples([(60, 75), (75, 80), (80, 95)])
df['weight_cut'] = pd.cut(df['weight'],bins)
print(df)
```

```python
df = pd.DataFrame(data={'weight': [75., 78.5, 85., 91., 84.5, 83., 68.]})
df['weight_cut'] = pd.cut(
    df['weight'],
    bins=(60, 75, 80, 95),
    labels=['light', 'normal', 'heavy'],
)
print(df)
```
