```python
# Solution 1
A = np.array(['1', '2', '3'], dtype=str)
print(np.char.add('000', A))
```

```python
# Solution 2
A = np.array(['1', '2', '3'], dtype=str)
print(np.char.rjust(A, 4, fillchar='0'))
```

```python
# Solution 3
A = np.array(['1', '2', '3'], dtype=str)
print(np.char.zfill(A, 4))
```
