```python
import pandas as pd

stocks = {'PLW': 387.00, 'CDR': 339.5, 'TEN': 349.5, '11B': 391.0}
quotations = pd.Series(data=stocks)
```

```python
quotations=pd.DataFrame(quotations)
quotations.columns = ['price']
print(quotations)
```

```python
a1 = np.arange(101,110)
a2 = np.arange(10,100,10)
Series = pd.Series(a2,a1,dtype=float)
print(Series)
```
