```python
import pandas as pd
import numpy as np

predictions = pd.read_csv("predictions.csv")

def mean_absolute_error(y_true,y_pred):
    return abs(y_true - y_pred).sum()/len(y_true)

mae = mean_absolute_error(predictions['y_true'],predictions['y_pred'])
print(mae)
```

```python
import pandas as pd
import numpy as np

predictions = pd.read_csv("predictions.csv")

def mean_squared_error(y_true,y_pred):
    diff = y_true - y_pred
    diff_2 = diff*diff
    return diff_2.sum()/len(y_true)

mse = mean_squared_error(predictions['y_true'],predictions['y_pred'])
print(f"MSE = {round(mse,4)}")
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)
df = pd.DataFrame(data=np.random.randn(10), columns=['var1'])
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

df['var1_sigmoid'] = df['var1'].apply(sigmoid)
print(df)
```
