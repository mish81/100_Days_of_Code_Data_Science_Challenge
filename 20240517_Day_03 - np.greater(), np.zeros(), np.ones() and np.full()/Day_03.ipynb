Basic Usage

np.greater(x1, x2) performs element-wise comparison between arrays x1 and x2.

Parameters

x1, x2: Input arrays to be compared. They should have the same shape or be broadcastable to a common shape.

```python
import numpy as np
```

```python
A = np.array([0.4, 0.5, 0.3, 0.9])
B = np.array([0.38, 0.51, 0.3, 0.91])

print(np.greater(A,B))
```

Basic Usage

np.zeros(shape, dtype=float, order='C')

Parameters

- shape: The shape of the array. It can be an integer or a tuple of integers specifying the dimensions of the array.

- dtype (optional): The data type of the array. Default is float.

- order (optional): Whether to store multi-dimensional data in row-major ('C', default) or column-major ('F') order in memory.

```python
print(np.zeros([4,4],dtype=int))
```

Basic Usage

np.ones(shape, dtype=None, order='C')

Parameters
- shape: The shape of the array. It can be an integer or a tuple of integers specifying the dimensions of the array.
- dtype (optional): The data type of the array. If not specified, the default dtype is float64.
- order (optional): Whether to store multi-dimensional data in row-major ('C', default) or column-major ('F') order in memory.

```python
print(255*np.ones([10,10]))
```

Basic Usage

np.full(shape, fill_value, dtype=None, order='C')

Parameters
- shape: The shape of the new array. This can be an integer or a tuple of integers specifying the dimensions of the array.
- fill_value: The constant value to fill the array with.
- dtype (optional): The data type of the array. If not specified, NumPy will determine the data type based on the fill_value.
- order (optional): Whether to store multi-dimensional data in row-major ('C', default) or column-major ('F') order in memory.

```python
print(np.full([10,10],255.))
```
