```python
import pandas as pd
```

```python
amazon = pd.read_csv('amazon.csv',index_col=0)
tesla = pd.read_csv('tesla.csv',index_col=0)
```

```python
print(amazon)
print()
print(tesla)
```

```python
amazon_columns = list(amazon.columns)
tesla_columns = list(tesla.columns)

for i in range(len(amazon_columns)):
    amazon_columns[i] = 'amzn_' + amazon_columns[i].lower()
amazon_columns

for i in range(len(tesla_columns)):
    tesla_columns[i] = 'tsla_' + tesla_columns[i].lower()

amazon.columns = amazon_columns
tesla.columns = tesla_columns

print(amazon)
print()
print(tesla)
```

```python
result = pd.concat([amazon,tesla],axis=1)
print(result)
```
