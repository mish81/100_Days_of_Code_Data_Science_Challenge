```python
import pandas as pd


url = (
    'https://storage.googleapis.com/esmartdata-courses-files/'
    'ml-course/insurance.csv'
)
df = pd.read_csv(url)
```

```python
df[['sex','smoker','region']] = df[['sex','smoker','region']].astype('category')
df.info()
```

```python
df_int = df.select_dtypes(['float','int'])
print(list(df_int.columns))
```

```python
df_cat = df.select_dtypes('category')
df_num  = df.select_dtypes(['float','int'])

print(df_cat.head())
print(df_num.head())
```
