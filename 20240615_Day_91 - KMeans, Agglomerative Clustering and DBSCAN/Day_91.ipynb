```python
import numpy as np
import pandas as pd
import random
from numpy.linalg import norm

# Assuming df is your DataFrame and has columns x1 and x2
# Example DataFrame creation for demonstration:
# df = pd.DataFrame({'x1': np.random.rand(100), 'x2': np.random.rand(100)})

x1_min = df["x1"].min()
x1_max = df["x1"].max()

x2_min = df["x2"].min()
x2_max = df["x2"].max()

centroid_1 = np.array(
    [
        random.uniform(x1_min, x1_max),
        random.uniform(x2_min, x2_max)
    ]
)

centroid_2 = np.array(
    [
        random.uniform(x1_min, x1_max),
        random.uniform(x2_min, x2_max)
    ]
)

for i in range(10):
    clusters = []
    for point in df[['x1', 'x2']].values:
        centroid_1_dist = norm(centroid_1 - point)
        centroid_2_dist = norm(centroid_2 - point)
        cluster = 1
        if centroid_1_dist > centroid_2_dist:
            cluster = 2
        clusters.append(cluster)
    
    df['cluster'] = clusters
    
    print(clusters)
    
    centroid_1 = np.array([
        round(df[df['cluster'] == 1]['x1'].mean(), 3),
        round(df[df['cluster'] == 1]['x2'].mean(), 3)
    ])

    centroid_2 = np.array([
        round(df[df['cluster'] == 2]['x1'].mean(), 3),
        round(df[df['cluster'] == 2]['x2'].mean(), 3)
    ])

print("Final centroids:")
print("Centroid 1:", centroid_1)
print("Centroid 2:", centroid_2)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

np.random.seed(42)
```

```python
df = pd.read_csv("data_q2.csv")
kmeans = KMeans(n_clusters = 3, max_iter=1000,random_state=42)
kmeans.fit(df)

print(kmeans.cluster_centers_)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

np.random.seed(42)
df = pd.read_csv('data_q3.csv')

kmeans = KMeans(n_clusters=3, max_iter=1000, random_state=42)
kmeans.fit(df)
```

```python
y_kmeans = kmeans.predict(df)
df['y_kmeans'] = y_kmeans
print(df.head(10))
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

np.random.seed(42)
df = pd.read_csv('data_q4.csv')
 
wcss = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df)
    wcss.append(round(kmeans.inertia_, 2))
print(wcss)
```
