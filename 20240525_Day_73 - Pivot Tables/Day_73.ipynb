```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 30)
pd.set_option('display.width', 1000)
tips = pd.read_csv(
    'https://raw.githubusercontent.com/mwaskom/'
    'seaborn-data/master/tips.csv'
)
```

```python
tips.head()
```

```python
print(pd.pivot_table(tips,values=['tip','total_bill'],index=['day'],aggfunc={
    'tip': np.mean,
    'total_bill': np.mean
}))
```

```python
pivot_data = pd.pivot_table(tips, 
                values = ['tip','total_bill'],
                index = ['day','sex'],
                aggfunc={
                    'tip' : np.mean,
                    'total_bill' : np.mean
                })

print(pivot_data)
```

```python
tips = pd.read_csv(
    'https://raw.githubusercontent.com/mwaskom/'
    'seaborn-data/master/tips.csv'
)
```

```python
data = pd.pivot_table(tips,
                values = ['tip','total_bill'],
                index = ['day'],
                aggfunc = {
                    'tip':max,
                    'total_bill':max
                })

print(data)
```

```python
data = pd.pivot_table(tips,
               values = ['tip','total_bill'],
               index = ['day','sex','smoker'],
               aggfunc = {
                'tip':np.mean,
                'total_bill':np.mean 
               }
                )

print(data)
```
